### Experiments on Digits using PCA

##### Student comment for below code: <br> 
* To begin with, we start with a simple approach on the X array. We take the average of the all pixel values. The result is meaningless and unusable, as seen from plotting. Because, we expect it to be at least similar to any number to can be used as model.
* Actually this is a data preparation process. We will try to reduce the data storage.

```python
plt.matshow(np.mean(X, axis=0).reshape((8, 8))) 
plt.title('Average of all digits')
```

Average of all digits is useless.

##### Student comment for below code: <br> 
* We should select necessary ones from this columns to classify numbers. So, we decided to use PCA (Principal Component Analysis). Generally, PCA used for <b>dimensionality reduction</b> in machine learning. Als it is an unsupervised, non-parametric statistical technique.
* In here we are dealing with a high dimentional problem. The term "high dimensionality" refers to a dataset with a large number of features. Model overfitting, which limits the ability to generalize beyond the examples in the training set, is the fundamental issue related with high-dimensionality in the machine learning area.
* We will now get more accurate results with the fitting process we have done with the PCA method.

```python
from sklearn.decomposition import PCA

pca = PCA()
scores = pca.fit_transform(X)  
```

##### Student comment for below code: <br> 
* The blue line shows the <b>cumulative sum</b>. Cumulative sum is sum of current and all previous values.
* The explained variance ratio is an array of the variance of the data explained by each of the principal components in our data.
* As shown in the figure below after <b>12 components more than 80%</b> of the total variance is captured.
* Our use of PCA, in the sense that only the dominating digits in the data come to the surface.
* If the mean of all pixels were belong to correct number in a high resolution, we would capture all the variability in one major component, that is, <b>we would capture the 100% explained variance after the first component (eigenvector) below</b>. From this we can conclude that the data matrix <b>X contains many different</b> pixel values that belongs to different numbers.
* The plot below clearly shows that most of the variance can be explained by the first 12 component alone. The principal that after 12th component until 24th or 25th component still bears some information while after 30th principal components can safely be dropped without losing too much information. Together, the first 12 principal components contain big amount of the information.

```python
plt.figure(figsize = (10, 8))
plt.plot(pca.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = '--') 
plt.title('Explained Variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid() 
```

### Experiments on each Digits using PCA

##### Student comment for below code: <br> 
* In here, unlike above one, we have separately trained each class instead of the whole data. Obviously we expect better result on fitting. Because with this approach, in earlier components, we reach better cumulative explained variance result for each class.  
* Since we took the mean of each class's pixel values here, we encountered a much better result than the result after taking the mean of all the data like in the previous example, as can be seen in the figures below.

```python
meanDigits = []
digitPCAs = [] 

plt.gray()
plt.figure(figsize=(15, 15))
for index in digits:
    indices = (y == index)
    Xd = data.data[indices]
    
    meanDigit = np.mean(Xd, axis=0)    
    plt.subplot(1, 10, index+1).matshow(meanDigit.reshape((8, 8)))  
    meanDigits.append(meanDigit)    
    
    pca = PCA()
    scores = pca.fit_transform(Xd)  
    digitPCAs.append(pca)
```

##### Student comment for below code: <br> 
* As expected we can easily obtain that less than 10 components more than 80% of the total variance is captured which is a better result comparing to PCA on all the data. So, for each class less amount of eigenvectors contain most of the information comparing the eigenvectors for all the data.
* In here each figure shows the cumulative explained variance result according to components (eigenvectors) for unique digits from 0 to 9 so there are 10 figures.

```python
for index in digits:
    pca = digitPCAs[index]
    
    plt.figure(figsize = (10, 8))
    plt.plot(pca.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = '--') 
    plt.title('Explained Variance by Components')
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.grid()  
```

Let take only first 16 components over 64 components (eigenvectors), i.e. reduced to %25 of the original storage while only 10% information is lost

##### Student comment for below code: <br> 
* The plot above plots clearly shows that most of the variance can be explained by the less components (eigenvectors). So, we decided to drop these unnecassary eigenvectors, which are contain small amount of information, in this way we decrease storage usage. We decided to use first 16 eigenvectors. These eigenvectors contain much more information comparing the after 16th to 64th eigenvector.
* As can be seen from the images, the numbers are very clear in the first pictures, as the first eigenvector contains the most information. However, the sharpness decreases towards the final images, because the 16th eigenvector contains less information than the 1st eigenvector.
* As a result, here we see how each eigen vector produces a result one by one.

```python
rank = 16

i = 1
plt.gray()
plt.figure(figsize=(15, 15))
for index in digits:
    pca = digitPCAs[index]
    components = pca.components_[0:rank, :]
    
    # mean digit
    meanDigit = meanDigits[index]
    plt.subplot(digits.size, rank+1, i).matshow(meanDigit.reshape((8, 8))) 
    i = i + 1
    
    # first 16 eigenvectors (reshaped from vector of 64 to 8x8 to show as an image)
    for r in range(0, rank):
        plt.subplot(digits.size, rank+1, i).matshow(components[r, :].reshape((8, 8))) 
        i = i + 1
   
```

##### Student comment for below code: <br> 
* Here we use only the eigen vector, which contains the most information at first. Then, we see the result when we add other eigen vectors one by one in order of importance. For instance, second pictures of each digits consist of 1st and 2nd eigen vector. Actually, the clarity in the 1st images is not bad because it was created with the 1st eigen vector, which contains most of the information. But when we put other eigen vectors on top of the previous one, we can observe that the result can get better and better. So, finally, we use all of the first 16 eigen vectors in the last  pictures. In the previous image, we used each eigen vector alone.

```python
i = 1    
plt.gray()
plt.figure(figsize=(15, 15))    
for digit in digits:
    meanDigit = meanDigits[digit]
    pca = digitPCAs[digit]
    components = pca.components_[0:rank, :]

    sv = pca.singular_values_[0:rank]
    
    for rk in range(0, rank):
        rg = np.random.normal(0, 1, size=(rank, 1))
        randomDigit = meanDigit;
        for r in range(0, rank): 
            randomDigit = randomDigit + rg[r]*np.sqrt(sv[r])*components[r, :]   

        plt.subplot(digits.size, rank, i).matshow(randomDigit.reshape((8, 8))) 
        i = i + 1
```
