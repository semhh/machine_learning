# <u>Forward Selection</u>
Algorithmic Approach

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
import statsmodels.api as sm

def forward_selection(data, target, significance_level=1/10**20):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<significance_level):
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features
```

In the method of <u>Forward Selection</u>, a mapping is used in which no feature is used at the beginning. Then, their performance is measured using the train-test split method as an example for each feature. Our <u>initial mapping</u> value was M0 = { }. Let the <u>next one</u> be M1 = { } and if we say <u>x5</u> for the feature with the best performance, now <u>M1 = {x5}</u>. We then measure the performance of each map by adding the best performing features to a new map, <u>one by one</u>. We use the featurs with the best performance, i.e. the <u>best accuracy</u> or the <u>lowest R^2</u> value, to select the fitting suitable for the data. Also, in order for us to perform Forward Selection, the number of data(n) does not need to be greater than dimention(p) as in Backward Selection. It can be used in such cases. We use Forward Selection for data with high dimentional. It may <u>not always produce the best subset</u>, but considering the <u>compilation efficiency</u>, this method is preferred.

# <u>Lasso Regression</u>
Mathematical Approach

```python
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model

# Find LASSO paths (LASSO coeffections for different alpha values)
n_alphas = 50
lasso_alphas = np.linspace(1e-8, 100, n_alphas)

X_new = StandardScaler().fit_transform(X_b)

lasso_coefs = []
for a in lasso_alphas:
    lasso = linear_model.Lasso(alpha=a, fit_intercept=True, max_iter=10000)
    lasso.fit(X_new, y)
    lasso_coefs.append(lasso.coef_)
```

```python
# Show LASSO paths   
plt.figure(2, figsize=(6,4))    
plt.plot(lasso_alphas, lasso_coefs)
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('LASSO coefficients as a function of the regularization')
plt.xlim((0,100))
plt.ylim((-7.5,10))
plt.grid()
plt.show()
```

<u>As we can see we just need feature x2 and x6.</u>

```python
# LASSO regression for different alpha values
alphas = [0.01, 0.1, 10, 15, 25, 30, 45]
for a in alphas:
    lasso = linear_model.Lasso(alpha=a, fit_intercept=True, max_iter=10000)
    lasso.fit(X_new, y)

    print('LASSO coefficients for alpha =', a)
    print(lasso.coef_)
    print()
```
